```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
class NeuralNetwork(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.linear = nn.Linear(num_features, 1)
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, features):
        out = self.linear(features)
        out = self.sigmoid(out)
        return out

```

```python

epochs = 25

learning_rate = 0.1
loss = nn.BCELoss()
```

```python
model = NeuralNetwork(X_train_tensor.shape[1])
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)    
```

```python
# Ensure tensors are float32 for compatibility with model parameters
X_train_tensor = X_train_tensor.float()
y_train_tensor = y_train_tensor.float()

for epoch in range(epochs):
    y_pred = model(X_train_tensor) # forward pass
    # Reshape y_train_tensor to match y_pred shape
    loss_value = loss(y_pred, y_train_tensor.view(-1, 1)) # loss calculation
    loss_value.backward() # backward pass
    optimizer.step() # update weights
    optimizer.zero_grad()

    print(f'Epoch {epoch+1}/{epochs}, Loss: {loss_value.item()}')
```
